```python
# Bài làm được tham khảo từ code mẫu của Tutor Tuấn Bùi

import re
import statistics

# Task 1. Tạo hàm đọc file
def read_file():
    class_file = input("\nEnter a file name to grade: ")
    file = open(f"{class_file}.txt",'r')
    lines = file.read().split("\n")
    student_list = []
    for i in range(len(lines)):
        student_list.append(lines[i].split(","))
    print(f"\nSuccessfully opened: {class_file}.txt\n")
    print("****Analyzing****\n")
    # Trả về 2 biến class_file, student_list
    return class_file, student_list
# Hàm try/except được chuyển xuống dưới cùng bài để kiểm tra cùng 1 lúc 4 task
#try:
    #class_file, student_list = read_file()
#except:
    #print("File cannot be found")
    
# Task 2. Kiểm tra các dòng trong file dữ liệu hợp lệ theo yêu cầu
def check_valid(student_list):
    valid_lines = 0
    invalid_lines = 0
    valid_student_list = []
    # Tạo vòng lặp kiểm tra dữ liệu trong list
    for i in range(len(student_list)):
        if len(student_list[i])==26 and re.match(r"^N\d{8}$", student_list[i][0]):
            valid_lines +=1
            valid_student_list.append(student_list[i])
        elif len(student_list[i])!=26:
            invalid_lines +=1
            print(f"Invalid line of data: does not contain exactly 26 values:\n\n{','.join(student_list[i])}\n")
        elif not re.match(r"^N\d{8}$", student_list[i][0]):
            invalid_lines +=1
            print(f"Invalid line of data: N# is invalid:\n\n{','.join(student_list[i])}\n")
    if invalid_lines ==0:
        print("No errors found\n")
    print("**** REPORT ****\n")
    print(f"Total valid lines of data {valid_lines}\n")
    print(f"Total invalid lines of data {invalid_lines}\n")
    
    return valid_student_list
# Task3. Tính điểm thi
def grading(valid_student_list):
    # Biến đổi answer_key thành một list
    answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
    answer_list = answer_key.split(",")
    grades = []
    high_score_student = 0
    skip_question = []
    incorrect_question = []
    #Tạo các biến theo yêu cầu
    
    for i in range(len(valid_student_list)):
        grade = 0
        for j in range(1,len(valid_student_list[i])):
            if valid_student_list[i][j] == answer_list[j-1]:
                grade +=4
            elif valid_student_list[i][j] == "":
                grade +=0
                skip_question.append(j)
            else:
                grade -=1
                incorrect_question.append(j)
                
        if grade > 80:
            high_score_student +=1
        grades.append(grade)
    #Trả về danh sách điểm học sinh, số học sinh điểm cao, các câu được skip và các câu trả lời sai
    return grades, high_score_student, skip_question, incorrect_question
    print(skip_question)
def statistics_grades(grades, high_score_student, skip_question, incorrect_question, valid_student_list):
    # 3.1. Đếm số lượng học sinh đạt điểm cao (>80).
    print(f"Total student of high scores: {high_score_student}\n")
    # 3.2. Điểm trung bình.
    print(f"Average scores: {round(statistics.mean(grades),2)}\n")
    # 3.3. Điểm cao nhất.
    print(f"Max scores: {max(grades)}\n")
    # 3.4. Điểm thấp nhất.
    print(f"Min scores: {min(grades)}\n")
    # 3.5. Miền giá trị của điểm (cao nhất trừ thấp nhất).
    print(f"Range scores: {max(grades)-min(grades)}\n")
    # 3.6. Giá trị trung vị
    print(f"Average scores: {statistics.median(grades)}\n")
    
    #Tính toán các câu được skip nhiều nhẩt và tỉ lệ
    #3.7. Trả về các câu hỏi bị học sinh bỏ qua nhiều nhất theo thứ tự: số thứ tự câu hỏi - số lượng học sinh bỏ qua -  tỉ lệ bị bỏ qua
    most_skip_list = statistics.multimode(skip_question)
    print_skip_list = []
    for i in range(len(most_skip_list)):
        most_skip = most_skip_list[i]
        number_of_skip = skip_question.count(most_skip_list[i])
        skip_rate = round(skip_question.count(most_skip_list[i])/len(valid_student_list),2)
        print_skip_list.append(f"{most_skip} - {number_of_skip} - {skip_rate}")
    print(f"Question that most people skip: {', '.join(print_skip_list)}\n")
    
    #3.8. Trả về các câu hỏi bị học sinh sai qua nhiều nhất theo thứ tự: số thứ tự câu hỏi - số lượng học sinh trả lời sai - tỉ lệ bị sai
    most_incorrect_list = statistics.multimode(incorrect_question)
    print_most_incorrect = []
    for i in range(len(most_incorrect_list)):
        most_incorrect = most_incorrect_list[i]
        number_of_incorrect = skip_question.count(most_incorrect_list[i])
        incorrect_rate = round((skip_question.count(most_incorrect_list[i])/(len(valid_student_list))),2)
        print_most_incorrect.append(f"{most_incorrect} - {number_of_incorrect} - {incorrect_rate}")
    print(f"Question that most people skip: {', '.join(print_most_incorrect)}\n") 
        

def statistics_grades(grades, high_score_student, skip_question, incorrect_question, valid_student_list):
    #In ra các chỉ số thống kê về điểm học sinh trong lớp
    print(f"Total student of high scores: {high_score_student}\n")
    print(f"Mean (average) score: {round((statistics.mean(grades)),2)}\n")
    print(f"Higest score: {max(grades)}\n")
    print(f"Lowest score: {min(grades)}\n")
    print(f"Range of score: {max(grades)-min(grades)}\n")
    print(f"Median score: {statistics.median(grades)}\n")

    #Tính toán các câu được skip nhiều nhẩt và tỉ lệ
    most_skip_list = statistics.multimode(skip_question)
    print_most_skip = []
    for i in range(len(most_skip_list)):
        most_skip = most_skip_list[i]
        number_of_skip = skip_question.count(most_skip_list[i])
        skip_rate = round((skip_question.count(most_skip_list[i])/(len(valid_student_list))),2)
        print_most_skip.append(f"{most_skip} - {number_of_skip} - {skip_rate}")
    print(f"Question that most people skip: {', '.join(print_most_skip)}\n")

    #Tính toán các câu bị trả lời sai nhiều nhất và tỉ lệ
    most_incorrect_list = statistics.multimode(incorrect_question)
    print_most_incorrect = []
    for i in range(len(most_incorrect_list)):
        most_incorrect = most_incorrect_list[i]
        number_of_incorrect = skip_question.count(most_incorrect_list[i])
        incorrect_rate = round((skip_question.count(most_incorrect_list[i])/(len(valid_student_list))),2)
        print_most_incorrect.append(f"{most_incorrect} - {number_of_incorrect} - {incorrect_rate}")
    print(f"Question that most people skip: {', '.join(print_most_incorrect)}\n")

# Task4. Ghi lại kết quả tên class_grades.txt
def write_result(class_file, valid_student_list, grades):
    with open(f'{class_file}_grades.txt', 'w+') as writefile:
        for i in range(len(valid_student_list)):
            writefile.write(f"{valid_student_list[i][0]}, {grades[i]} \n")
    with open(f'{class_file}_grades.txt', 'r') as test_writefile:
        print(f"Nội dung của tệp sau khi ghi: {class_file}_grade.txt")
        print(test_writefile.read())        
try:
    class_file, student_list = read_file()
    valid_student_list = check_valid(student_list)
    grades, high_score_student, skip_question, incorrect_question = grading(valid_student_list)
    statistics_grades(grades, high_score_student, skip_question, incorrect_question, valid_student_list)
    write_result(class_file, valid_student_list, grades)
except:
    print("File cannot be found")
```
